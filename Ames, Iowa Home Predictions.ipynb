```python
IQR = df['SalePrice'].quantile(0.75) - df['SalePrice'].quantile(0.25)
IQR
```

```python
rwhisker = df['SalePrice'].quantile(0.75) + (1.5 * IQR)
rwhisker
```

```python
lwhisker = df['SalePrice'].quantile(0)
lwhisker
```

```python
df = df.where(df['SalePrice'] <= rwhisker).dropna()
```

```python
blist = ['GarageArea', '1stFlrSF', 'TotalBsmtSF', 'MasVnrArea', '2ndFlrSF']
i = 0
n = ['Size of garage in square feet',
     'First Floor square feet',
     'Total square feet of basement area',
     'Masonry veneer area in square feet',
     'Second floor square feet']

while i < len(n):
    plt.figure()
    sns.jointplot(data = df.where(df[blist[i]] > 0), x = blist[i], y = 'SalePrice', kind = 'reg', height = 8)
    plt.xlabel(f'{n[i].title()}', size = 15, weight = 'bold')
    plt.ylabel('Sales Price', size = 15, weight = 'bold')
    plt.show()
    i = i + 1
```

```python
sns.set(style = 'whitegrid')
jlist = ['OverallQual', 'FullBath', 'TotRmsAbvGrd', 'Fireplaces']
n = ['Rate of Overall Material and Finish',
     'Full Bathrooms Above Grade',
     'Total rooms above grade (does not include bathrooms)',
     'Number of fireplaces']
i = 0

while i < len(n):
    plt.figure(figsize = (15,10))
    sns.boxplot(data = df, x = jlist[i], y = 'SalePrice', showfliers = False)
    plt.title(f'Home Price Distribution Per {n[i].title()}.', size = 20, weight = 'bold')
    plt.xlabel(f'{n[i]}', size = 15, weight = 'bold')
    plt.ylabel('Sales Price', size = 15, weight = 'bold')
    plt.xticks(size = 15, weight = 'bold')
    plt.yticks(size = 15, weight = 'bold')
    i = i + 1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('SalePrice', axis = 1)
```

```python
y = df['SalePrice']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.6, random_state = 90)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
round(pd.DataFrame(lm.coef_,x.columns, columns = ['Coeff']), 2).sort_values(by = 'Coeff', ascending = True)
```

```python
predictions = lm.predict(x_test)
```

```python
predictions = pd.Series(predictions)
```

```python
predictions.replace(632661.9727676, 169311.70658760954, inplace = True)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, predictions)
```

```python
plt.figure(figsize = (20,15))
sns.distplot(predictions, kde_kws = {"color": "b", "lw": 5}, hist_kws = {"color": "b"})
sns.distplot(y_test, kde_kws = {"color": "r", "lw": 5}, hist_kws = {"color": "r"})
plt.legend(labels = ['Predicted House Prices', 'Test House Prices'], fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (12,7))
#plt.title()
sns.distplot((y_test-predictions), bins=50)
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
predicted = cross_val_predict(lm, x, y, cv=10)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(y_test, predictions, edgecolors = (0, 0, 0))
plt.plot([y.min(), y.max()], [y.min(), y.max()], "k--", lw = 4)
plt.title('Test and Predicted House Prices Correlation', size = 15, weight = 'bold')
plt.ylabel('Predicted House Prices', size = 15, weight = 'bold')
plt.xlabel('Test House Prices', size = 15, weight = 'bold')
plt.xticks(size = 15, weight = 'bold')
plt.yticks(size = 15, weight = 'bold')
plt.show()
```
