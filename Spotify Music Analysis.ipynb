```python
g = pd.read_csv('genres_v2.csv', low_memory=False)
g_df = pd.DataFrame(g)
```

```python
corr_d_df = g_df.corr()
```

__Please see heatmap to discover the correlations between differing attributes of music. \
The stronger the color, the stronger the correlation.__

```python
fig = plt.figure(figsize=(20,20))
x = sns.heatmap(corr_d_df, annot = True, linewidth = .5, linecolor = 'black', cmap = 'coolwarm')
x.set_ylim(sorted(x.get_xlim(), reverse=True))
plt.title('Correlations of Various Music Features', fontsize = 40)
plt.tight_layout()
plt.show()
```

__Learn of the distribution of frequency among the multiple attributes of music__

```python
fig, axes = plt.subplots(11, 1, figsize = (7,40))

ax1 = sns.distplot(g_df['danceability'], ax = axes[0], color = 'red')
ax1.title.set_text('Distribution of Danceability')
ax1.set_ylabel('Frequency of Rankings')
ax1.set_xlabel('Rankings')

ax2 = sns.distplot(g_df['energy'], ax = axes[1], color = 'cyan')
ax2.title.set_text('Distribution of Energy')
ax2.set_ylabel('Frequency of Rankings')
ax2.set_xlabel('Rankings')

ax3 = sns.distplot(g_df['key'], ax = axes[2], color = 'green')
ax3.title.set_text('Distribution of Key')
ax3.set_ylabel('Frequency of Rankings')
ax3.set_xlabel('Rankings')

ax4 = sns.distplot(g_df['loudness'], ax = axes[3], color = 'purple')
ax4.title.set_text('Distribution of Loudness')
ax4.set_ylabel('Frequency of Rankings')
ax4.set_xlabel('Rankings')

ax5 = sns.distplot(g_df['mode'], ax = axes[4], color = 'blue')
ax5.title.set_text('Distribution of Mode')
ax5.set_ylabel('Frequency of Rankings')
ax5.set_xlabel('Rankings')

ax6 = sns.distplot(g_df['speechiness'], ax = axes[5], color = 'yellow')
ax6.title.set_text('Distribution of Speechiness')
ax6.set_ylabel('Frequency of Rankings')
ax6.set_xlabel('Rankings')

ax7 = sns.distplot(g_df['acousticness'], ax = axes[6], color = 'black')
ax7.title.set_text('Distribution of Speechiness')
ax7.set_ylabel('Frequency of Rankings')
ax7.set_xlabel('Rankings')

ax8 = sns.distplot(g_df['instrumentalness'], ax = axes[7], color = 'orange')
ax8.title.set_text('Distribution of Instrumentalism')
ax8.set_ylabel('Frequency of Rankings')
ax8.set_xlabel('Rankings')

ax9 = sns.distplot(g_df['liveness'], ax = axes[8], color = 'violet')
ax9.title.set_text('Distribution of Liveness')
ax9.set_ylabel('Frequency of Rankings')
ax9.set_xlabel('Rankings')

ax10 = sns.distplot(g_df['valence'], ax = axes[9], color = 'magenta')
ax10.title.set_text('Distribution of Valence')
ax10.set_ylabel('Frequency of Rankings')
ax10.set_xlabel('Rankings')

ax11 = sns.distplot(g_df['tempo'], ax = axes[10], color = 'brown')
ax11.title.set_text('Distribution of Tempo')
ax11.set_ylabel('Frequency of Rankings')
ax11.set_xlabel('Rankings')

plt.tight_layout()

plt.show()
```
